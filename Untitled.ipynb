```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the spreadsheet data
df = pd.read_excel('spreadsheet.xlsx', engine='openpyxl')

# Perform analysis
summary = {
    'mean': df.mean().values[0],
    'median': df.median().values[0],
    'std_dev': df.std().values[0],
    'min': df.min().values[0],
    'max': df.max().values[0]
}

# Plot histogram
plt.figure(figsize=(10, 6))
sns.histplot(df, kde=True)
plt.title('Distribution of Values')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Plot boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title('Boxplot of Values')
plt.xlabel('Values')
plt.grid(True)
plt.show()

# Plot line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=df)
plt.title('Line Plot of Values')
plt.xlabel('Index')
plt.ylabel('Values')
plt.grid(True)
plt.show()

summary
```
